## **Exploratory Data Analysis (EDA)**

# 1. Quick Info

```python
print("Shape:", df.shape)
print("\nData Types:")
print(df.dtypes.value_counts())
print("\nFirst 5 rows:")
print(df.head())
```

# 2. Target Variable (Fraud Reported)

```python
if "fraud_reported" in df.columns:
    fraud_counts = df["fraud_reported"].value_counts()
    plt.figure(figsize=(5,4))
    fraud_counts.plot(kind="bar", color=["orange","green"], edgecolor="black")
    plt.title("Fraud Reported Distribution")
    plt.xlabel("Fraud Reported")
    plt.ylabel("Count")
    plt.show()
    print("Fraud distribution:\n", fraud_counts)
```

# 3. Numeric Values

```python
num_cols = df.select_dtypes(include=[np.number]).columns
print("Numeric Columns:", num_cols.tolist())

# Histograms
df[num_cols].hist(bins=20, figsize=(16,12), edgecolor="black")
plt.suptitle("Numeric Variables Histograms")
plt.show()

# Correlation heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df[num_cols].corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# 4. Categorical Variables

```python
cat_cols = df.select_dtypes(exclude=[np.number]).columns
print("Categorical Columns:", cat_cols.tolist())

# Top categories for first few categorical columns
for col in cat_cols[:5]:
    plt.figure(figsize=(8,4))
    sns.countplot(
        y=col,
        data=df,
        order=df[col].value_counts().index[:10],
        hue=df[col],       
        legend=False,      
        palette="muted"
    )
    plt.title(f"Top Categories for {col}")
    plt.show()
```

# 5. Feature vs Target (if fraud_reported exists)

```python
if "fraud_reported" in df.columns:
    # Example: Age vs Fraud
    if "age" in df.columns:
        plt.figure(figsize=(8,4))
        sns.boxplot(x="fraud_reported", y="age", data=df)
        plt.title("Age vs Fraud Reported")
        plt.show()

    # Example: Total Claim Amount vs Fraud
    if "total_claim_amount" in df.columns:
        plt.figure(figsize=(8,4))
        sns.boxplot(x="fraud_reported", y="total_claim_amount", data=df)
        plt.title("Total Claim Amount vs Fraud Reported")
        plt.show()
```
